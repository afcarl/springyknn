```python
from springyknn.springyknn import *

%matplotlib notebook
%load_ext autoreload
%autoreload 2
```

```python
from sklearn.datasets import load_wine
x_train, y_train = load_wine(True)
y_train = y_train.reshape(-1,1)
print(x_train.shape)
print(y_train.shape)
```

```python
springyknn = SpringyKNN(x_train, y_train, randomize=True, num_pivots=178)
G = springyknn.get_springy_knn_graph(k1=5, k2=5)
springyknn.write_gexf(G, filename='wine_springyknn_graph.gexf')
```

```python
springyknn.viz_spring_graph(G, dim=2, iterations=150, scale=1.0)
```
